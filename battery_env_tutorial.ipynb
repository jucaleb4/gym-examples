```python
import gymnasium as gym
import gym_examples

import numpy as np
import matplotlib.pyplot as plt
```

# BatteryEnv Tutorial
The goal of this notebook is to get familiar with the `BatteryEnv` environment. A couple things to note:

- There are 8640 data points in either the `real` or `periodic` dataset. This corresponds to 90 days worth of 15min intervals of LMPs and other data
- You can set the time limit length upon calling. Here, we set it to 2000 to allow for more time.

```python
nhistory = 10     # number of history points
data = "real" # get periodic (sinusoidal) data - algo "real"
env_mode = "default" # standard setting but delay cost of buying
start_index = 0
end_index = -1 # 24 * 4

env = gym.make(
    "gym_examples/BatteryEnv-v0", 
    pnode_id='FREMNT_1_N013',
    nhistory=nhistory, 
    data=data, 
    mode=env_mode, 
    start_index = start_index,
    end_index = end_index,
    more_data = True,
    delay_cost = False,
    max_episode_steps=2000, # can change length here!
)
```

Let's take a look at the state
- first index: SOC
- historical prices (length `nhistory`, working backwards in chronological order starting with the price right now)
- current "cost" of electricity (oldest unsold charge)

```python
for key in env.observation_space.keys():
    print(f"Key {key}: {env.observation_space[key]}")
```

Let's now look at action space.
- action=0 : sell
- action=1 : do nothing
- action=2 : buy

```python
print(env.action_space)
```

Let's test out buying, doing nothing, and then selling.

We'll also test out with some "invalid" actions, e.g., selling while empty or buying while full.

```python
s_0, _ = env.reset()
print(f"Initial state: {s_0}\n")
s_1, _, _, _, _ = env.step(2)
print(f"State after buying: {s_1}\n")
s_2, _, _, _, _ = env.step(1)
print(f"State after doing nothing: {s_2}\n")
s_3, _, _, _, _ = env.step(0)
print(f"State after selling: {s_3}\n")
s_4, _, _, _, _ = env.step(0)
print(f"State after selling with nothing: {s_4}\n")

print("buying a lot to fill up battery...")
for i in range(8):
    s_5, _, _, _, _ = env.step(2)
print(f"State with battery full: {s_5}\n")

s_6, _, _, _, _ = env.step(2)
print(f"State after buying with full battery : {s_6}\n")
```

```python
env = gym.make(
    "gym_examples/BatteryEnv-v0", 
    pnode_id = 'FREMNT_1_N013',
    nhistory=nhistory, 
    data=data, 
    daily_cost=1000,
    start_index = start_index,
    end_index = end_index,
    max_episode_steps=2000, # can change length here!
)
```

```python
battery_capacity = 400
transfer_rate = 50
daily_cost = 1000
env.reset()
# buy
for t in range(int(battery_capacity/transfer_rate)):
    s_i, r_i, _, _, _ = env.step(2)
    print(r_i, -transfer_rate*s_i["lmps"][1]-daily_cost)
# null
for t in range(int(battery_capacity/transfer_rate)):
    s_i, r_i, _, _, _ = env.step(1)
    print(r_i, -daily_cost)
# sell
for t in range(int(battery_capacity/transfer_rate)):
    s_i, r_i, _, _, _ = env.step(0)
    print(r_i, transfer_rate*s_i["lmps"][1]-daily_cost)
```

Let's visualize the data!

```python
num_days = 14
num_steps = num_days * 4 * 24

lmp_prices = np.zeros(num_steps)
s_0, _ = env.reset()
lmp_prices[0] = s_0["lmps"][0]

has_truncated = False
for t in range(1,num_steps):
    s, _, term, trun, _ = env.step(1)
    if trun and not has_truncated:
        print(f"Truncated starting at time {t}")
        has_truncated = True
    lmp_prices[t] = s["lmps"][0]

ax = plt.subplot()
ax.plot(np.arange(len(lmp_prices)), lmp_prices)
ax.set(
    title=f"{data} LMP prices over {num_days} days", 
    ylabel="Prices ($/MWh)", 
    xlabel="RT time steps"
)
```

## Adding solar

Let's now add solar and see when it peaks. We will focus on the last 14 days, where we typically do testing on.

```python
start_index = 4*24*(90-14)
end_index = 4*24*90

env = gym.make(
    "gym_examples/BatteryEnv-v0", 
    pnode_id = 'FREMNT_1_N013',
    nhistory=nhistory, 
    data=data, 
    mode=env_mode, 
    start_index = start_index,
    end_index = end_index,
    more_data = True,
    delay_cost = True,
    max_episode_steps=2000, # can change length here!
    solar_coloc=True,
    solar_scale=1.0,
)
```

```python
num_days = 14
num_steps = num_days * 24 * 4

solars = np.zeros(num_steps)
solar_fcsts = np.zeros(num_steps)
s_0, _ = env.reset()
solars[0] = s_0["solars"][0]
rescale = 0.01
solar_fcsts[0] = rescale * s_0["solar_fcsts"][0]

has_truncated = False
for t in range(1,num_steps):
    s, _, term, trun, _ = env.step(1)
    if trun and not has_truncated:
        print(f"Truncated starting at time {t}")
        has_truncated = True
    if term:
        print(f"Term starting at time {t}")
    solars[t] = s["solars"][0]
    solar_fcsts[t] = rescale * s["solar_fcsts"][0]

ax = plt.subplot()
ax.plot(np.arange(len(solars)), solars, color="blue", label="solar")
ax.plot(np.arange(len(solar_fcsts[4:])), solar_fcsts[4:], color="red", label="fcst")
ax.vlines(np.arange(14)*24*4, ymin=0, ymax=25, color="black")
ax.set(
    title=f"{data} solar over {num_days} days (vertical lines are new days)", 
    ylabel="Power (MWh)", 
    xlabel="RT time steps"
)
ax.legend()

print("Max occurs at time delta %s" % np.argmax(solars))
print(solars[4*6:])
```
